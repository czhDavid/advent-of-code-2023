The newly-improved calibration document consists of lines of text; each line originally contained a specific calibration value that the Elves now need to recover. On each line, the calibration value can be found by combining the first digit and the last digit (in that order) to form a single two-digit number.

For example:

1abc2
pqr3stu8vwx
a1b2c3d4e5f
treb7uchet
In this example, the calibration values of these four lines are 12, 38, 15, and 77. Adding these together produces 142.

```python
def decodeLine(codeword: str) -> int:
    firstNumber = ""
    secondNumber = ""
    for letter in codeword[::]:
        if (letter.isnumeric()):
            firstNumber = letter
            break;

    for letter in codeword[::-1]:
        if (letter.isnumeric()):
            secondNumber = letter
            break;   
    return int(firstNumber + secondNumber)

def decodeLines(codewords: [int]):
    total = 0
    for codeword in codewords:
        total += decodeLine(codeword)
        
    return total
    
```

```python
def findFirstLetterInCodeword(codeword: str) -> str:
    numericNumberDictionary = {'one': 1,
                           'two': 2,
                           'three': 3,
                           'four': 4,
                           'five': 5,
                           'six': 6,
                           'seven': 7,
                           'eight': 8,
                          'nine': 9 }
    
    lowestPosition = len(codeword) + 1
    number = ''
    for numberStr in list(numericNumberDictionary.keys()):
        index = codeword.find(numberStr)
        if (index != -1 and lowestPosition > index):
            lowestPosition = index
            number = str(numericNumberDictionary[numberStr])
            
    for index in range(len(codeword)):
        if (codeword[index].isnumeric()) and lowestPosition > index:
            return codeword[index]
        
    return number
def findLastLetterInCodeword(codeword: str) -> str:
    numericNumberDictionary = {'one': 1,
                           'two': 2,
                           'three': 3,
                           'four': 4,
                           'five': 5,
                           'six': 6,
                           'seven': 7,
                           'eight': 8,
                          'nine': 9 }
    
    highestPosition = -1
    number = -1
    for numberStr in list(numericNumberDictionary.keys()):
        index = codeword.rfind(numberStr)
        if (index != -1 and highestPosition < index):
            highestPosition = index
            number = str(numericNumberDictionary[numberStr])
            
    for index in range(len(codeword))[::-1]:
        if (codeword[index].isnumeric()) and highestPosition < index:
            return codeword[index]
        
    return number
```

```python
elfInput = open("elf_input.txt", "r")
total = 0
for line in elfInput:
    print(line)
    print(str(findFirstLetterInCodeword(line)) + str(findLastLetterInCodeword(line)))
    total += int(str(findFirstLetterInCodeword(line)) + str(findLastLetterInCodeword(line)))
print(total)

elfInput.close()
```
