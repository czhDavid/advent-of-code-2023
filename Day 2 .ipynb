```python
def is_possible(game_input, red_cubes, green_cubes, blue_cubes):
    for subset in game_input:
        cubes = subset.split(', ')
        red = 0
        green = 0
        blue = 0
        for cube in cubes:
            count, color = cube.split()
            count = int(count)
            if color == 'red':
                red += count
            elif color == 'green':
                green += count
            elif color == 'blue':
                blue += count

        if red > red_cubes or green > green_cubes or blue > blue_cubes:
            return False

    return True

def possible_games(input_data, red_cubes, green_cubes, blue_cubes):
    possible_games_list = []
    for line in input_data:
        game_id, subsets = line.split(': ')
        game_id = int(game_id.split()[-1])
        subsets = subsets.split('; ')
        if is_possible(subsets, red_cubes, green_cubes, blue_cubes):
            possible_games_list.append(game_id)

    return possible_games_list

# Your input data (games information)
# games_input = [
#     "Game 1: 3 blue, 4 red; 1 red, 2 green, 6 blue; 2 green",
#     "Game 2: 1 blue, 2 green; 3 green, 4 blue, 1 red; 1 green, 1 blue",
#     "Game 3: 8 green, 6 blue, 20 red; 5 blue, 4 red, 13 green; 5 green, 1 red",
#     "Game 4: 1 green, 3 red, 6 blue; 3 green, 6 red; 3 green, 15 blue, 14 red",
#     "Game 5: 6 red, 1 blue, 3 green; 2 blue, 1 red, 2 green"
# ]

# Define the cube counts
red_count = 12
green_count = 13
blue_count = 14

# Get possible games
possible_games_result = possible_games(games_input, red_count, green_count, blue_count)
print("Games that would have been possible:", possible_games_result)
print("Sum of IDs of possible games:", sum(possible_games_result))
```
