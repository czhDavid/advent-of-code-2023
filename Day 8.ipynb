This format defines each node of the network individually. For example:

RL

AAA = (BBB, CCC)
BBB = (DDD, EEE)
CCC = (ZZZ, GGG)
DDD = (DDD, DDD)
EEE = (EEE, EEE)
GGG = (GGG, GGG)
ZZZ = (ZZZ, ZZZ)
Starting with AAA, you need to look up the next element based on the next left/right instruction in your input. In this example, start with AAA and go right (R) by choosing the right element of AAA, CCC. Then, L means to choose the left element of CCC, ZZZ. By following the left/right instructions, you reach ZZZ in 2 steps.

Of course, you might not find ZZZ right away. If you run out of left/right instructions, repeat the whole sequence of instructions as necessary: RL really means RLRLRLRLRLRLRLRL... and so on. For example, here is a situation that takes 6 steps to reach ZZZ:

LLR

AAA = (BBB, BBB)
BBB = (AAA, ZZZ)
ZZZ = (ZZZ, ZZZ)
Starting at AAA, follow the left/right instructions. How many steps are required to reach ZZZ?



```python
import re
from math import lcm
```

```python
with open('day8.txt') as fileInput: 
    instructions, nodes = fileInput.read().split('\n\n')
```

```python
nodesDict = {}
for node in nodes.split('\n'):
    i,a,b = re.findall('(\w+) = \((\w+)+, (\w+)\)',node)[0]
    nodesDict[i] = (a,b)
```

```python
currentPosition = 'AAA'
howManySteps = 0
while currentPosition != 'ZZZ': 
    for side in instructions:
        howManySteps+=1
        indexToGo = 0 if side == 'L' else 1
        currentPosition = nodesDict[currentPosition][indexToGo]
print(howManySteps)
```

```python
def getStartingPossitions(nodeKeys): 
    startingPostitions=[]
    for nodeKey in nodeKeys: 
        if nodeKey[-1] == 'A':
            startingPostitions.append(nodeKey)
    return startingPostitions
```

```python
startingPositions = getStartingPossitions(nodesDict.keys())
```

```python

currentPositions = startingPositions
positionCycleLength = []

for currentPositionIndex in range(len(currentPositions)):

    currentPosition = currentPositions[currentPositionIndex]
    howManySteps = 0
    while currentPosition[-1] != 'Z': 
        for side in instructions:
            howManySteps+=1
            indexToGo = 0 if side == 'L' else 1
            currentPosition = nodesDict[currentPosition][indexToGo]
    positionCycleLength.append(howManySteps)

```

```python
print(positionCycleLength)
print(lcm(*positionCycleLength))
```
