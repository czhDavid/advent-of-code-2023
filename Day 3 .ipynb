```python
def processEngineLine(lines:[str])->list[str]:
    lineToCheck = lines[1]
    foundNumbers= re.findall('\d+',lineToCheck)
    foundEngineParts = []
    characterPosition = 0
    while characterPosition < len(lineToCheck):
        if lineToCheck[characterPosition].isnumeric():
            foundNumber= re.search('\d+',lineToCheck[characterPosition:])[0]
            print(foundNumber)
            positionsToCheckForEachLine = createPositionsToCheckForEachLine(characterPosition, characterPosition -1 + len(foundNumber))
            if isEnginePart(positionsToCheckForEachLine, lines):
                foundEngineParts.append(foundNumber)
            characterPosition+=len(foundNumber)
            continue
        characterPosition +=1
    return foundEngineParts
    
```

```python
def findPositionsOnLine(number: str, line: str) -> int:
    searchStartPoint=0
    while True:
        i = line.index(number, searchStartPoint)
        if (not line[i-1].isnumeric() or i==0) and not line[min(i + len(number), len(line)-1)].isnumeric():
            return i
        searchStartPoint=i +1
```

```python
def createPositionsToCheckForEachLine(numberStart: int,numberEnd: int)-> list[int]:
    positionsToCheckInOtherLines=[numberStart - 1,numberEnd + 1] + [x + numberStart for x in range(numberEnd-numberStart + 1)]
    linePostionsToCheck = [
        positionsToCheckInOtherLines,
        [numberStart - 1,numberEnd+1],
        positionsToCheckInOtherLines
    ]
    return linePostionsToCheck
```

```python
def isEnginePart(positionsToCheckForEachLine, lines) -> bool:
    for lineIndex in range(len(lines)):
        currentLine = lines[lineIndex]
        for positionToCheck in positionsToCheckForEachLine[lineIndex]:
            if positionToCheck < 0 or positionToCheck > LINE_LENGTH-1:
                continue
            if currentLine[positionToCheck] !='.' and not currentLine[positionToCheck].isnumeric():
                #print(lineIndex, positionToCheck)
                return True
    return False
```

```python
def processEngineLineForCogs(lines:[str]):
    lineToCheck = lines[1]
    characterPosition = 0
    offsetsToCheckForNumberOnOtherLinesRelativeToCog = [-1,0,1]
    offsetsToCheckOnSameLine = [-1,1]
    
    cogRatio = 0

    for positionIndex in range(len(lineToCheck)):
        if lineToCheck[positionIndex] == '*':
            cogRatio += findGearsForCogs(lines, positionIndex)
            
    return cogRatio
```

```python
def findGearsForCogs(lines:[str], center: int) -> int:
    positionsOffsetsToCheck = [-1,0,1]
    gears = []
    for line in lines:
        partToCheck = line[max(center -1,0): min(center + 2, len(line)-1)]
        matches = re.search(r"\.?([0-9]+)\D?([0-9]+)?", partToCheck)
        if matches == None:
            continue
        matchedGroups = [x for x in matches.groups() if x != None]
        for matchIndex in range(1, len(matchedGroups)+1):
            gears.append(getNumberOnPosition(line,center -1 + matches.start(matchIndex)))
        if len(gears) ==2:
            return gears[0] * gears[1]
        
    return 0
```

```python
def getNumberOnPosition(line: str,position: int) -> int:
    #find out where the number ends
    regexp = re.compile(r"([0-9]+)")
    #start goes from zero
    for match in regexp.finditer(line):
        if position >= match.start() and position < match.end():
            return int(match.group())
```
